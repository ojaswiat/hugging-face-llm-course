## NLP
- Enables computers to understand, interpret, and generate human language.
- Encompasses many techniques and tasks such as sentiment analysis, named entity recognition, and machine translation.

### Common NLP Tasks
- **Classifying whole sentences:** Sentiment analysis, spam detection, grammar checking, sentence correlations.
- **Classifying each word in a sentence:** Identify grammatical components - nouns, pronouns, adjectives, verbs, etc.
- **Generating text content:** Prompt completion, conversation, etc.
- **Extracting an answer from a text:** Answer synthesis based on training data.
- **Generating a new sentence from an input text:** Translating a text into another language, summarizing a text, etc.

## LLMs
- Subset of NLP models.
- Characterized by their massive size, extensive training data, and ability to perform a wide range of language tasks with minimal task-specific training. 
- Models like the Llama, GPT, or Claude series are examples of LLMs that have revolutionized what’s possible in NLP.

A large language model (LLM) is an AI model trained on massive amounts of text data that can understand and generate human-like text, recognize patterns in language, and perform a wide variety of language tasks without task-specific training. They represent a significant advancement in the field of natural language processing (NLP).

### LLM Parameters
- Parameters are the number of weights and biases (neurons).
- They act as the "general context" and "memory (storage)".
- A 70B parameter LLM is better than a 7B parameter LLM because it has more "memory".

### LLM Characteristics
- **Scale:** They contain millions, billions, or even hundreds of billions of parameters.
- **General capabilities:** They can perform multiple tasks without task-specific training.
- **In-context learning:** They can learn from examples provided in the prompt.
- **Emergent abilities:** As these models grow in size, they demonstrate capabilities that weren’t explicitly programmed or anticipated.

### LLM Limitations
- **Hallucinations:** They can generate incorrect information confidently.
- **Lack of true understanding:** They lack true understanding of the world and operate purely on statistical patterns.
- **Bias:** They may reproduce biases present in their training data or inputs.
- **Context windows:** They have limited context windows (though this is improving).
- **Computational resources:** They require significant computational resources.

### Why Language Processing is Challenging?
LLMs use mathematical concepts to model the natural language **approximately, not exactly** using mathematics. Hence the limitations.

## Transformers and Pipelines

Transformer models are used to solve all kinds of tasks across different modalities, including natural language processing (NLP), computer vision, audio processing, and more.

### The `pipeline()` Function

- The most basic object in the transformers library.
- The transformer expects numbers (vectors) and not raw text.
- Handles all the pre-processing and post-processing steps.

### How the `pipeline()` Function Works

- Creates a `pipeline` object and selects a particular pretrained model that has been fine-tuned for the purpose.
- The model is downloaded and cached when creating the pipeline object. No need to download the model multiple times.
- Steps done by the pipeline object:
    - The text is preprocessed into a format the model can understand.
    - The preprocessed inputs are passed to the model.
    - The predictions of the model are post-processed, so you can make sense of them.


### Available Pipelines
[View all the pipelines by HuggingFace in the Transformers docs here.](https://huggingface.co/docs/hub/en/models-tasks)

#### Text pipelines
- sentiment-analysis: Analyse sentiments from a text.
- text-generation: Generate text from a prompt.
- text-classification: Classify text into predefined categories.
- summarization: Create a shorter version of a text while preserving key information.
- translation: Translate text from one language to another.
- zero-shot-classification: Classify text without prior training on specific labels.
- feature-extraction: Extract vector representations of text.

#### Image pipelines
- image-to-text: Generate text descriptions of images.
- image-classification: Identify objects in an image.
- object-detection: Locate and identify objects in images.

#### Audio pipelines
- automatic-speech-recognition: Convert speech to text.
- audio-classification: Classify audio into categories.
- text-to-speech: Convert text to spoken audio.

#### Multimodal pipelines
- image-text-to-text: Respond to an image based on a text prompt.


### Import the `pipeline` Module

1. Takes a fair amount of time to import.
2. Takes a fair amount of time to download and cache the appropriate model for the task.

```python
# Import the pipeline object from the transformers library
import torch
from transformers import pipeline

# Import the pprint function for pretty printing
from pprint import pprint as pretty_print

# Set the default model for the pipelines
DEFAULT_MODEL = "HuggingFaceTB/SmolLM2-135M" # A Smol Model
```

### 1. Sentiment Analysis
- Uses a classifier to classify sentiments - positive and negative.
- Give scores to each of the sentiments representing the strength.

```python
# Using the sentiment analysis pipeline to classify the sentiment of a given text

classifier = pipeline("sentiment-analysis", model=DEFAULT_MODEL) # Create a sentiment analysis pipeline
sentiment = classifier("I've been waiting for a HuggingFace course my whole life.") # Analyze the sentiment of the input text

pretty_print(sentiment) # Print the raw sentiment output
print()
print(f"Sentiment: {sentiment[0]['label']}, Confidence: {sentiment[0]['score']:.2f}") # Print the sentiment label and confidence score
```

```python
# Pass multiple sentences to the classifier
sentences = [
    "I've been waiting for a HuggingFace course my whole life.",
    "I hate this movie.",
    "This is the best day of my life!"
]

sentiments = classifier(sentences) # Analyze the sentiment of multiple sentences

for sentiment, sentence in zip(sentiments, sentences):
    print(f"Sentence: {sentence}")
    print(f"Sentiment: {sentiment['label']}, Confidence: {sentiment['score']:.2f}")
    print() # Print a newline for better readability
```

### 2. Zero Shot Classification
- Works on unlabelled data.
- Classify the texts based on custom labels with confidence scores.

```python
# Using a zero-shot classification pipeline to classify a sentence into one of several candidate labels

classifier = pipeline("zero-shot-classification", model=DEFAULT_MODEL) # Create a zero-shot classification pipeline

sentence = "This is a course about the Transformers library." # Analyze the sentence using zero-shot classification
labels = ["education", "politics", "sports"] # Define candidate labels for classification

result = classifier(sentence, candidate_labels=labels) # Classify the sentence into one of the candidate labels

pretty_print(result) # Print the raw result of the classifications
print()
print(f"Predicted label: {result['labels'][0]}, Confidence: {result['scores'][0]:.2f}") # Print the predicted label and confidence score
```

```python
# Multiple texts for zero-shot classification
sentences = [
    "This is a course about the Transformers library.",
    "The government has announced new policies.",
    "The football match was exciting."
]
labels = ["education", "politics", "sports"] # Define candidate labels for classification

results = classifier(sentences, candidate_labels=labels) # Classify multiple sentences into one of the candidate labels

for result in results:
    print(f"Sentence: {result['sequence']}")
    pretty_print(result) # Print the raw result of the classifications
    print(f"Predicted label: {result['labels'][0]}, Confidence: {result['scores'][0]:.2f}") # Print the predicted label and confidence score
    print() # Print a newline for better readability
```

### 3. Text Generation
- Uses predictive text like keyboard in phone.

```python
generator = pipeline("text-generation", model=DEFAULT_MODEL) # Specify the model for text generation

sequence = "In this course, we will teach you how to"

result = generator(
    sequence,
    max_length=30,
    num_return_sequences=2,
)

pretty_print(result) # Print the raw result of the text generation
print()
print(f"Generated text: {result[0]['generated_text']}") # Print the generated text
```

### 4. Mask Filling
- Fill the `<blanks>` in a given text

```python
unmasker = pipeline("fill-mask", model="distilbert/distilroberta-base") # Specify the model for masked token filling, DEFAULT_MODEL not supported for this task

masked_sequence = "The capital of France is <mask>."

results = unmasker(masked_sequence, top_k=2) # Fill the masked token in the sequence with the top 2 predictions

pretty_print(results) # Print the raw result of the masked token filling
print()
for result in results:
    print(f"Predicted token: {result['token_str']}, Confidence: {result['score']:.2f}") # Print the predicted token and confidence score
    print() # Print a newline for better readability
```
